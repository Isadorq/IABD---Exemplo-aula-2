```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
Renda = np.array([1200, 5000, 8000, 
                  10000, 30000]).reshape(-1, 1)
Gastos = np.array([750, 1500, 4000,
                   6000, 18000])

modelo = LinearRegression()
modelo.fit(Renda, Gastos)
```

```python
G_pred = modelo.predict(Renda)
```

```python
plt.scatter(Renda, Gastos, color='blue', label='Dados Reais')  
plt.plot(Renda, G_pred, color='green', label='Linha de Regressão')  

plt.xlabel('Renda')
plt.ylabel('Gastos')
plt.title('Previsão de gastos')
plt.legend()
plt.grid(True)
plt.show()
```
