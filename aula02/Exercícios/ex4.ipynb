```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('monitoramento_tempo.csv')
```

```python
df.head()
```

```python
df['data'] = pd.to_datetime(df['data'])  
```

```python

max_vento = df['direção do vento'].max()  # Valor máximo de umidade
min_vento = df['direção do vento'].min()  # Valor mínimo de umidade


max_data = df['data'][df['direção do vento'].idxmax()]
min_data = df['data'][df['direção do vento'].idxmin()]

```

```python
plt.figure(figsize=(10,6)) 
plt.plot( df['data'],df['direção do vento'], color="purple") 
plt.title('Gráfico de direção do vento', fontsize=15)

plt.annotate(f'Máximo: {max_vento}%', xy=(max_data, max_vento), xytext=(max_data, max_vento), fontsize=12, color='red')
plt.annotate(f'Mínimo: {min_vento}%', xy=(min_data, min_vento), xytext=(min_data, min_vento), fontsize=12, color='green')

plt.xlabel('Ano')
plt.ylabel('direção do vento')
```
