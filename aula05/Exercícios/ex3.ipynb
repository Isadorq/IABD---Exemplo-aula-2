```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
A = np.array([50, 100, 120, 300]).reshape(-1, 1)
Valor = np.array([180.000, 300.000, 
                  375.000, 600.000])

modelo = LinearRegression()
modelo.fit(A, Valor)
```

```python
V_pred = modelo.predict(A)
```

```python
plt.scatter(A, Valor, color='purple', label='Dados Reais')  
plt.plot(A, V_pred, color='pink', label='Linha de Regressão')  

plt.xlabel('A')
plt.ylabel('Valor')
plt.title('Previsão de valor de imóvel')
plt.legend()
plt.grid(True)
plt.show()
```
