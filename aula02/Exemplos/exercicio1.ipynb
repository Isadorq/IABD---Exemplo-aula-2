```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
# Cria dicionarios dados1 e dados2
dados1 = {
    "Data": ["01-01-2023","01-02-2023", "01-03-2023"],
    "Produto": ["Item A", "Item B", "Item A"],
    "Quantidade": [10, 5, 8],
    "Valor": [100.00, 50.00, 80.00]
}

dados2 = {
    "Data": ["01-02-2023", "02-02-2023", "03-02-2023"],
    "Produto": ["Item C", "Item B", "Item D"],
    "Quantidade": [12, 6, 3],
    "Valor": [120.00, 60.00, 30.00]
}

df1 = pd.DataFrame(dados1)
df2 = pd.DataFrame(dados2)
```

```python
df1.head()
```

```python
df_n = pd.concat([df1, df2]) # Junta os dois dataframes
```

```python
df_n.head(6)
```

```python
plt.bar(df_n['Produto'], df_n['Quantidade'], color = 'purple') # altera a cor do grafico 
plt.figure(figsize=(12, 12)) # altera o tamanho do grafico
plt.xlabel("Nome do Produto")   # adiciona um título no eixo
plt.ylabel("Quantidade")    # adiciona um título no eixo
plt.title("Produtos x Quantidade de Vendas")    # adiciona um título no gráfico
plt.legend(loc="upper right")   # adiciona uma legenda no gráfico
```
