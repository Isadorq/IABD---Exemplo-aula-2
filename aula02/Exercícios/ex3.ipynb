```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('monitoramento_tempo.csv')
```

```python
df.head()
```

```python
df['data'] = pd.to_datetime(df['data'])  
```

```python

max_humidade = df['humidade'].max()  # Valor máximo de umidade
min_humidade = df['humidade'].min()  # Valor mínimo de umidade


max_data = df['data'][df['humidade'].idxmax()]
min_data = df['data'][df['humidade'].idxmin()]

```

```python
plt.figure(figsize=(6,6)) 
plt.plot( df['data'],df['humidade'], color="pink") 
plt.title('Gráfico de Umidade', fontsize=15)

plt.annotate(f'Máximo: {max_humidade}%', xy=(max_data, max_humidade), xytext=(max_data, max_humidade), fontsize=12, color='red')
plt.annotate(f'Mínimo: {min_humidade}%', xy=(min_data, min_humidade), xytext=(min_data, min_humidade), fontsize=12, color='green')

plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
```
