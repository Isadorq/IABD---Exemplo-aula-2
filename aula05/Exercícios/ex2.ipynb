```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
Q = np.array([100, 200,300,400,500,600,700]).reshape(-1, 1)
T = np.array([30,35,40,45,60,100,150])
modelo = LinearRegression()
modelo.fit(Q, T)
```

```python
T_pred = modelo.predict(Q)
```

```python
plt.scatter(Q, T, color='blue', label='Dados Reais')  
plt.plot(Q, T_pred, color='red', label='Linha de Regressão')  

plt.xlabel('Q')
plt.ylabel('T')
plt.title('Previsão da temperatura do forno')
plt.legend()
plt.grid(True)
plt.show()
```
