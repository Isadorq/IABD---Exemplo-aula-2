```python
import pandas as pd # importa a biblioteca pandas
```

```python
df = pd.read_csv('imigrantes_canada.csv')   # importa a base de dados
```

```python
df.head()   # exibe as informações do dataframe
```

```python
df.info()   # exibe as informações do dataframe 
```

```python
df.set_index(['País'], inplace=True)    # seta o index com a coluna País
anos = list(map(str, range(1980, 2013)))    # converte os anos em string
# armazenados em uma lista
```

```python
brasil = df.loc['Brasil', anos] # pega os dados somente para o país
```

```python
brasil_dic = {
    'ano': brasil.index.tolist(),   # transforma a coluna com o index em lista
    'imigrantes': brasil.values.tolist()    # transforma a coluna com os valores em lista
}
```

```python
df_brasil = pd.DataFrame(brasil_dic)    # criando dataframe
```

```python
df_brasil   # exibe os valores
```

```python
df_brasil.info()    # exibe informações sobre o novo dataframe criado
```

```python
import matplotlib.pyplot as plt     # importa a biblioteca matplotlib
```

```python
plt.plot(df_brasil['ano'], df_brasil['imigrantes']) # plota o gráfico
plt.xticks(['1980', '1985', '1995', '2000', '2005', '2010'])    #ajuste de 
plt.yticks([500, 1000, 1500, 2000, 2500, 3000]) # ajuste os valores
plt.xlabel("Anos", fontsize=14) #titulo eixo x
plt.ylabel("Número de imigrantes", fontsize=14) # titulo eixo y
plt.show()  # exibe a figura
```

```python
china = df.loc['China', anos] # pega os dados somente para o país
```

```python
china_dic = {
    'ano': china.index.tolist(),   # transforma a coluna com o index em lista
    'imigrantes': china.values.tolist()    # transforma a coluna com os valores em lista
}
```

```python
df_china = pd.DataFrame(china_dic)  
```

```python
plt.plot(df_china['ano'], df_china['imigrantes']) # plota o gráfico
plt.xticks(['1980', '1985', '1995', '2000', '2005', '2010'])    #ajuste de 
plt.yticks([500, 1000, 1500, 2000, 2500, 3000]) # ajuste os valores
plt.xlabel("Anos", fontsize=14) #titulo eixo x
plt.ylabel("Número de imigrantes", fontsize=14) # titulo eixo y
plt.show()  # exibe a figura
```

```python
df_comparacao = df.loc[['Brasil', 'China', 'Cuba'], anos]   # pega duas colunas Brasil e China
```

```python
cuba = df.loc['Cuba', anos] 
```

```python
cuba_dic = {
    'ano': cuba.index.tolist(),   # transforma a coluna com o index em lista
    'imigrantes': cuba.values.tolist()    # transforma a coluna com os valores em lista
}
```

```python
df_cuba = pd.DataFrame(cuba_dic)  
```

```python
plt.plot(df_cuba['ano'], df_cuba['imigrantes']) # plota o gráfico
plt.xticks(['1980', '1985', '1995', '2000', '2005', '2010'])    #ajuste de 
plt.yticks([500, 1000, 1500, 2000, 2500, 3000]) # ajuste os valores
plt.xlabel("Anos", fontsize=14) #titulo eixo x
plt.ylabel("Número de imigrantes", fontsize=14) # titulo eixo y
plt.show()  # exibe a figura
```

```python
df_comparacao = df_comparacao.T # matriz transposta
```

```python
df_comparacao.head()
```

```python
plt.plot(df_comparacao['Brasil'], label= 'Brasil')
plt.plot(df_comparacao['China'], label= 'China')
plt.plot(df_comparacao['Cuba'], label= 'Cuba')
plt.title("Comparação de imigrantes do Brasil para a China \n e de Cuba para o Brasil")
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))   # criando uma figura
# ax - axes
ax.plot(df_brasil['ano'], df_brasil['imigrantes'], lw=3, color = 'green')   #comando
ax.xaxis.set_major_locator(plt.MultipleLocator(5))  # altera o intervalo do eixo x
ax.xaxis.set_tick_params(labelsize=12)
ax.xaxis.set_tick_params(labelsize=12)
ax.grid(linestyle='--')
ax.set_title('Imigração do Brasil para a China \n 1980 a 2013', loc='left')
ax.set_xlabel('Ano', fontsize=14)
ax.set_ylabel('Nº de imigrantes', fontsize=14)
plt.show()
```

```python
# criando a figura
fig, axs = plt.subplots(1, 2, figsize=(15,5))

# subplot 1: Gráfico de Linhas
axs[0].plot(df_brasil['ano'], df_brasil['imigrantes'])
axs[0].set_title("Imigração do Brasil para Cuba n\ 1980 a 2013")
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Número de Imigrantes')

axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()
# Subplot 2: Boxplot
axs[1].boxplot(df_brasil['imigrantes'])
axs[1].set_title('Box-plot da imigração d Brasil para China \n 1980 a 2013')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Cuba')
axs[1].grid()
```
