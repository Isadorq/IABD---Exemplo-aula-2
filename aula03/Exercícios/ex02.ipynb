```python
import pandas as pd
import matplotlib.pyplot as plt 
import datetime 
```

```python
df = pd.read_csv('monitoramento_tempo.csv')
```

```python
df['data'] = pd.to_datetime(df['data'])
```

```python
plt.plot(df['velocidade do vento'], df['data'])
plt.title("Vento x Ano", fontsize=10)
plt.xlabel("Vento")
plt.ylabel("Ano")
plt.figure(figsize=(10, 10))
plt.show()
```
