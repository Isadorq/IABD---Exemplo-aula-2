Umidade X Ano

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('monitoramento_tempo.csv')
```

```python
df.head()
```

```python
df['data'] = pd.to_datetime(df['data'])  
```

```python
plt.figure(figsize=(6,6)) 
plt.plot( df['humidade'], df['data'], color="purple")  
plt.title('Umidade X ano', fontsize=15)
plt.xlabel('Ano')
plt.ylabel('Umidade')
```
