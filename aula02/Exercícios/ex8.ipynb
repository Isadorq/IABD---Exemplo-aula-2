```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('monitoramento_tempo.csv')
```

```python
df.head()
```

```python
df['data'] = pd.to_datetime(df['data'])  
```

```python
plt.figure(figsize=(10,6)) 
plt.plot( df['data'],df['pressão'], color="purple") 
plt.title('Gráfico de Pressão', fontsize=15)
plt.xlabel('Ano')
plt.ylabel('Pressão')
```
