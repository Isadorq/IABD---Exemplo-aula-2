```python
import numpy as np
from sklearn.model_selection import train_test_split   # importa uma biblioteca de teste
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import accuracy_score, precision_score, confusion_matrix
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
data = pd.DataFrame({
    'consumo': [1000, 3000, 1500, 2500, 1200, 3500],
    'tempo': [200, 500, 300, 400, 250, 600],
    'tipo': ['eletrica', 'gas', 'eletrica', 'gas', 'eletrica', 'gas'],
    'classificacao': ['eficiente', 'ineficiente', 'eficiente', 'ineficiente', 'eficiente', 'ineficiente']
})
```

```python
x = np.array(data['consumo']) 
y = np.array(data['classificacao'])   
x_treino, x_teste, y_treino, y_teste = train_test_split(x.reshape(-1, 1), y.reshape(-1, 1))
model = DecisionTreeClassifier()
model.fit(x_treino, y_treino)
prev= model.predict(x_teste)
acuracy = accuracy_score(y_teste, prev)
matrix = confusion_matrix(y_teste, prev)
print(acuracy, matrix)
fig = plt.figure(figsize=(10,8))
tree.plot_tree(model, feature_names=x.tolist(), class_names=y.tolist(), filled=True)
plt.show()
```
