```python
import pandas as pd # importa a biblioteca pandas
import matplotlib.pyplot as plt # importa a biblioteca matplotlib
import numpy as np # importa a biblioteca numpy
```

```python
# Gerando n√∫meros com a Numpy
dados_x = np.linspace(0, 10, 10)    # gera valores coma  biblioteca numpy
dados_y = dados_x * dados_x # gera valores multiplicando um dado pelo 
```

```python
plt.figure(figsize=(10, 5), facecolor='gray')
Ax = plt.axes() # cria os eixos do gr√°fico 
Ax.set_facecolor('white')   # cria o fundo do gr√°fico
plt.plot(dados_x, dados_y,
         color='pink', linewidth=2,
         marker='o', linestyle='dashed', markeredgecolor='black')
plt.title("ExemploMatplotlib")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
```

```python
plt.rcParams.update({'font.size':15})   # define o tamanho da fonte
plt.figure(figsize=(12, 12))  # define o tamanho da figura
linha = 2   # define o tamanho de linhas
colunas = 2     # define o numero de colunas
plt.subplot(linha, colunas, 1)    # define o n√∫mero de linhas e colunas
plt.plot(dados_x, 'r--')
plt.title("Gr√°fico 1")  
 # grafico 2
plt.subplot(linha, colunas, 2)
plt.plot(dados_x * dados_x, 'b-.')
plt.title("Gr√°fico 2")

 # gr√°fico 3
plt.subplot(linha, colunas, 3)
plt.plot(dados_x/2, 'c-.')
plt.title("Gr√°fico 3")

# gr√°fico 4
plt.subplot(linha, colunas, 4)
plt.plot(dados_y*2)
plt.title("Gr√°fico 4")

plt.savefig("Exemplo subplot.jpg", dpi=300, transparent=True,
            bbox_inches='tight')    # comando para salvar a figura
```

```python
# Gr√°fico de barras
figura = plt.figure()   # armazena na vari√°vel Figura a figura a ser gerada 
grafico = figura.add_axes([0, 0, 0.7, 0.7]) # cria a grade do gr√°fico
rotulos = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']    # cria os r√≥tulos
grafico.bar(rotulos, dados_x, color='pink')
grafico.bar(rotulos, dados_x*0.1, color='gray')
plt.title("Quantidade x Categoria")
plt.xlabel("Itens")
plt.ylabel("Qualidade")
```
