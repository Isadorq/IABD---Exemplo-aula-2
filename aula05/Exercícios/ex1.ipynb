```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
vazao = np.array([10, 20, 30, 5, 35, 40]).reshape(-1, 1)

nivel = vazao * 0.2

modelo = LinearRegression()
modelo.fit(vazao, nivel)
```

```python
n_pred = modelo.predict(vazao)
```

```python
plt.scatter(vazao, nivel, color='orange', label='Dados Reais')  
plt.plot(vazao, n_pred, color='blue', label='Linha de Regressão')  

plt.xlabel('Vazão')
plt.ylabel('Nível')
plt.title('Previsão de vazão da água')
plt.legend()
plt.grid(True)
plt.show()
```
