```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
Consumo = np.array([50, 140, 30, 75, 123]).reshape(-1, 1)
Demanda = np.array([100, 200, 80,
                    100, 160])

modelo = LinearRegression()
modelo.fit(Consumo, Demanda)
```

```python
d_pred = modelo.predict(Consumo)
```

```python
plt.scatter(Consumo, Demanda, color='green', label='Dados Reais')  
plt.plot(Consumo, d_pred, color='gray', label='Linha de Regressão')  

plt.xlabel('Consumo')
plt.ylabel('Demanda')
plt.title('Previsão demanda de energia')
plt.legend()
plt.grid(True)
plt.show()
```
